```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" , attrs={"class": "wikitable sortable"})

print(tmp)
```

```python
#Вивести перші рядки таблиці за допомогою методу head
print(tmp[0].head(1))
```

```python
#Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
print(tmp[0].shape)
```

```python
#Замініть у таблиці значення "—" на значення NaN
tmp[0].fillna("-", inplace=True)
print(tmp[0])
```

```python
#Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—" (метод to_numeric)
tmp[0].replace("-", 0, inplace=True)
print(tmp[0])
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
tmp_1 = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" , attrs={"class": "wikitable sortable"})
# print(tmp_1[0])
tmp_null =tmp_1[0].isnull().sum()/len(tmp_1[0])*100
print(tmp_null)
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці (метод drop)
print(tmp[0].drop(tmp[0].index[-1], inplace=True))
print(tmp[0])
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
tmp_2 = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" , attrs={"class": "wikitable sortable"},index_col=0 )
print(tmp_2[0])
new_tmp = tmp_2[0].fillna(tmp_2[0].mean(), axis=0)
print(new_tmp)
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні (метод loc)
print(tmp_2[0].loc["Київ", "2014"])
print(f'Середній в  Україні за 2014 рік: {tmp_2[0].loc["Україна","2014"].mean()}')
print(f"Середній в  Україні за всі роки {tmp_2[0].loc['Україна'].mean()}")

print(tmp_2[0].loc[tmp_2[0]["2014"] > tmp_2[0].loc["Україна"].mean()].index)
```

```python
#У якому регіоні була найвища народжуваність у 2014 році? (метод idxmax)
print(tmp_2[0].loc[tmp_2[0]["2014"].idxmax()].name)
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2014 році (метод plot)
tmp_2[0].plot(kind="bar",
                y="2014",
                figsize=(20, 10),
                title="Народжуваність по регіонах України у 2014 році",
                xlabel="Регіони",
                ylabel="Народжуваність",
                rot=90,
                #grid=True,
                legend=True,
                fontsize=7,
                colormap="coolwarm",
                edgecolor="black",
                linewidth=1,
                stacked=True,
                alpha=0.7,
                subplots=True,
                sharex=True,
                sharey=True,
                layout=(2, 2)
              )
plt.show()
```

```python
#Візуалізація даних з використанням бібліотеки matplotlib
data = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96" , attrs={"class": "wikitable sortable"}, index_col=0)

x = data[0].loc["Україна"].index
y = data[0].loc["Україна"].values
plt.plot(x, y)
plt.title("Народжуваність в Україні")
plt.xlabel("Роки")
plt.ylabel("Кількість народжених")
plt.grid(True)
plt.show()
```

```python
#Візуалізація даних з використанням бібліотеки seaborn
x_1 = data[0].loc["Київ"].index
y_1 = data[0].loc["Київ"].values
sns.set_style("darkgrid")

sns.lineplot(data=data[0].loc["Київ"],
            # x=x_1,
            # y=y_1,
            color="g",
            label="Київ",
            marker="D",
            markersize=10,
            )
plt.show()
```

```python
# Вибираємо 2 регіони та візуалізуємо їх народжуваність за роками
x_1 = data[0].loc["Київ"].index
y_1 = data[0].loc["Київ"].values
x_2 = data[0].loc["Вінницька область"].index
y_2 = data[0].loc["Вінницька область"].values
y_3 = data[0].loc["Україна"].values

#використовуємо бібліотеку seaborn
sns.set_style("darkgrid")
data_set = pd.DataFrame({"Київ": y_1, "Вінницька": y_2, "Україна":y_3, "Роки":x_2 }, index=x_1)
sns.lineplot(
            data=data_set,
            palette="ocean",
            linewidth=2.5,
            markers=True,
            legend="full",
             )
xlabel = plt.xlabel("Роки")
ylabel = plt.ylabel("Кількість народжених")
plt.show()
```
