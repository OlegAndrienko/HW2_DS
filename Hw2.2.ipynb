```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
url = "https://raw.githubusercontent.com/OlegAndrienko/HW2_DS/main/2017_jun_final.csv"
data_1 = pd.read_csv(url)
print(data_1.head(5))
print(data_1.groupby(["Должность"]))
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада",
# виконайте агрегацію даних за допомогою методу agg
# і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

position_data = data_1.groupby(["Должность"]).agg({"Зарплата.в.месяц": ["min", "max"]})
print(position_data)
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
# Використовуйте її для методу apply та створіть новий стовпчик "avg"

def fill_avg_salary(salary: pd.Series):
    return salary.mean()

avg = []
size_data = len(data_1)

for i in range(size_data):
    avg.append(fill_avg_salary(data_1["Зарплата.в.месяц"]))

data_1["avg"] = avg
print(data_1.head(5))
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика
print(data_1["avg"].describe())
```

```python
#Збережіть отриману таблицю в CSV файлі за допомогою методу to_csv
data_1.to_csv("new_data.csv", index=False)
```

```python
url = "https://raw.githubusercontent.com/OlegAndrienko/HW2_DS/main/2017_jun_final.csv"
data = pd.read_csv(url)

#формуємо датасет для візуалізації Должность - Зарплата.в.месяц
position_data = data.groupby(["Должность"]).agg({"Зарплата.в.месяц": ["min", "max"]})
min_salary = position_data["Зарплата.в.месяц"]["min"]
max_salary = position_data["Зарплата.в.месяц"]["max"]
# position = data.loc["Должность"]
# print(position)
data_set = pd.DataFrame({"min": min_salary, "max": max_salary,  },  index=position_data.index)
print(data_set)

#Візуалізація даних з використанням бібліотеки seaborn гістограма
sns.set_style("darkgrid")
sns.histplot(data=data_set,
             x="min",
             y="Должность",

   )

plt.show()
```
